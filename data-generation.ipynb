```python
import csv
import math

rows = []
```

Каждый "ряд" - набор данных, список рядов затем в csv передадим

```python
def traffic_intensity(row):
   return row[0] / row[1]

def prob_free_system(row):
  return 1 - traffic_intensity(row)

def n_prob(row, n):
  return (traffic_intensity(row) ** n) * prob_free_system(row)

def less_then_one_prob(row):
  return prob_free_system(row) + n_prob(row, 1)

def middle_num_of_system_clients(row):
  return row[0] / (row[1] - row[0])

def middle_num_of_queue_clients(row):
  return row[0] * row[0] / (row[1] * (row[1] - row[0]))

def middle_time_of_system_clients(row):
  return 1 / (row[1] - row[0])

def middle_time_of_queue_clients(row):
  return row[0] / (row[1] * (row[1] - row[0]))

def len_more_or_is_k_prob(row):
  return traffic_intensity(row) ** row[2]

def len_more_then_k_prob(row):
  return traffic_intensity(row) ** (row[2] + 1)

def queue_time_more_then_t(row):
  print(f"deg={-(row[1] - row[0]) / row[3]}")
  return math.exp(-(row[1] - row[0]) / row[3])

def system_time_more_then_t(row):
  return queue_time_more_then_t(row) * traffic_intensity(row)

def middle_queue_len(row):
  return row[1] * middle_time_of_system_clients(row)

def ten_clients_prob(row):
  return (1 - traffic_intensity(row)) * (traffic_intensity(row) ** 10)
```

Реализовали аналитические функции для "выходов"

```python
for l in range(60, 120): # 60 sec - 2 days
    for m in range(30, 720): # 30 sec - 2 hours
        for k in range(5, 10):
            for t in range(10 * m, 20 * m + 1):
                for n in range(10, 100):
                  row = [l, m, k, t, n]
                  analytical_values = [traffic_intensity(row),
                                       prob_free_system(row),
                                       n_prob(row, row[-1]),
                                       less_then_one_prob(row),
                                       middle_num_of_system_clients(row),
                                       middle_num_of_queue_clients(row),
                                       middle_time_of_system_clients(row),
                                       middle_time_of_queue_clients(row),
                                       len_more_or_is_k_prob(row),
                                       len_more_then_k_prob(row),
                                       queue_time_more_then_t(row),
                                       system_time_more_then_t(row),
                                       middle_queue_len(row),
                                       ten_clients_prob(row)]
                  final_row = [0] * (5 + 14 + 14)
                  ind = 0
                  for elem in row:
                    final_row[ind] += elem
                    ind += 1
                  for i in range(2): # сначала пусть идут аналитические значения, а потом - "реальные", то есть зашумленные аналитические
                    for elem in analytical_values:
                      final_row[ind] += elem
                      ind += 1
                  rows.append(final_row)
```
